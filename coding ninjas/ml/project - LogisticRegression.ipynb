```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")
df.head()
```

```python
def getSex(x):
    if x=="female":
        return 0
    else:
        return 1
def is_nan(x):
    return (x != x)

def getOrd(c):
    if is_nan(c):
        return c
    return ord(str(c))
def transformCabin(cabins):
    if is_nan(cabins):
        return cabins
    cabin_list = cabins.split(" ")
    c_int = 0
    for c in cabin_list:
#         print(c,c[1:])
        c_int+=ord(c[0].strip())
        for c1 in c[1:].strip().split(" "):
            if c1.strip() == '':
                continue
            c_int+=int(c[1:].strip())
    return c_int
```

```python
data_x = df.drop("Name",axis=1)
data_x = data_x.drop("Ticket",axis=1)
data_x = data_x.drop("Cabin",axis=1)
data_x = data_x.drop("Survived",axis=1)
# data_x["Sex"] = data_x.Sex.map({'female':0, 'male':1})
data_x["Sex"] = data_x.Sex.apply(getSex)
data_x["Embarked"] = data_x.Embarked.apply(getOrd)
# data_x["Cabin"] = data_x.Cabin.apply(transformCabin)
data_x.Embarked.fillna(data_x.Embarked.mean(),inplace=True)
data_x.Age.fillna(data_x.Age.mean(),inplace=True)
# data_x.Cabin.fillna(data_x.Cabin.mean(),inplace=True)
data_x
# data_x["Cabin"] = data_x.Cabin.replace("B96","")
# for i in data_x.Embarked:
#     data_x[""]
data_y = df["Survived"]
data_y
data_x
```

```python
# clf = LogisticRegression()
clf = LogisticRegression(solver='newton-cg',multi_class="ovr",max_iter=1000)
clf.fit(data_x,data_y)
```

```python
# clf.predict((cancer_ds.data))
clf.score(data_x,data_y)
```
