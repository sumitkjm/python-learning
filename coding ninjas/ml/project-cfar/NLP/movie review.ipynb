```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
documents[0:5]
```

```python
import random
random.shuffle(documents)
documents[0:5]
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
import string
punctuations = string.punctuation
stop = stop + list(punctuations)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk import pos_tag
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents_new = [(clean_review(doc),cat) for doc, cat in documents]
```

```python
documents_clean = documents_new
```

```python
training_documents = documents_clean[0:1500]
testing_documents = documents_clean[1500:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
def get_feature_dict(words,features):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
training_data = [ (get_feature_dict(doc,features),cat) for doc, cat in training_documents]
testing_data = [ (get_feature_dict(doc,features),cat) for doc, cat in testing_documents]
```

```python
training_data[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
cls = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(cls,testing_data)
```

```python
nltk.pred(testing_data)
```

```python
cls.show_most_informative_features(15)
```
