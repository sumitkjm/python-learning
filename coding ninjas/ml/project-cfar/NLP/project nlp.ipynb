```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
import string
punctuations = string.punctuation
stop = stop + list(punctuations)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
def clean_words(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
train_docs = [(clean_words(word_tokenize(df_value[7])),df_value[1]) for df_value in df_train.values]
test_docs = [clean_words(word_tokenize(df_value[6])) for df_value in df_test.values]
test_docs
```

```python
all_train_words = []
for doc in train_docs:
    all_train_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_train_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
def get_feature_dict(words,features):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```
