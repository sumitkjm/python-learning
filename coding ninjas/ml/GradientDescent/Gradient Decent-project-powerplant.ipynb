```python
import numpy as np
from sklearn import preprocessing
```

```python
def step_gradient(points, learning_rate,m):
    m_slope = [0 for i in range(len(m))]
    M = len(points)
    for i in range(len(points)):
        y = points[i,len(m)]
        temp_m = 0
        for j in range(len(m)):
            x = points[i,j]
            temp_m+=m[j]*x
        for j in range(len(m)):
            x = points[i,j]
            m_slope[j]+=(-2/M) * (y-temp_m)*x
    new_m = [0 for i in range(len(m))]
    for j in range(len(m)):
        new_m[j] = m[j] - learning_rate*m_slope[j]
    return new_m
```

```python
def gd(points,learning_rate,num_iterations,nf):
    # start with random value of m, c
    m = [0 for i in range(nf+1)]
    for i in range(num_iterations):
        m = step_gradient(points,learning_rate,m)
        print(i, " Cost: ", cost(points,m))
    return m
```

```python
def cost(points, m):
    M = len(points)
    total_cost = 0
    for i in range(M):
        y = points[i,len(m)]
        cost_temp = 0
        for j in range(len(m)):
            x = points[i,j]
            cost_temp += m[j]*x
        total_cost += (1/M)*((y-cost_temp)**2)
    return total_cost
```

```python
def scale(points,scale):
    scale = preprocessing.StandardScale()
    scale.fit(points)
    return points
```

```python
def pred(test_points, m):
    M = len(m)
    Y = [0 for i in range(len(test_points))]
    for i in range(len(test_points)):
        temp_y = 0
        for j in range(M-1):
            temp_y +=test_points[i,j]*m[j]
        temp_y+=m[M-1]
        Y[i] = temp_y
    return Y
```
