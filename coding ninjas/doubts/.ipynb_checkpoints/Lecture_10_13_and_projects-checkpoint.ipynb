```python
import csv
import numpy as np
with open("terrorismData.csv") as file_:
    file_name=csv.DictReader(file_)
    country=[]
    kill=[]
    wound=[]
    group=[]
    for i in file_name:
        if (i["Year"]=="1999"):
            if (i["Country"]=="India" and i["State"]=="Jammu and Kashmir"):
                if (int(i["Month"])>=5 and int(i["Month"])<8):
                    if (int(i["Day"])>0):
                        country.append(i["City"])
                        group.append(i["Group"])
                        if i["Killed"]=="":
                            kill.append("0.0")
                        else:
                            kill.append(i["Killed"])
                        if i["Wounded"]=="":
                            wound.append("0.0")
                        else:
                            wound.append(i["Wounded"])
    country=np.array(country)
    kill=np.array(kill,dtype=float)
    wound=np.array(wound,dtype=float)
    group=np.array(group)
    kill=kill+wound
    l=np.argmax(kill)
    print(int(kill[l]),country[l],group[l])
```

```python
import csv
import numpy as np
with open("terrorismData.csv", "r") as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace = True)
    casualty = []
    city = []
    for row in file_data:
        if row["Year"] == "1999" and (row["Month"] == "5" or row["Month"] == "6" or row["Month"] == "7"):
            if row["State"] == "Jammu and Kashmir":
                casualty.append((row["Killed"], row["Wounded"]))
                city.append((row["City"], row["Group"]))
    np_casualty = np.array(casualty)
    np_city = np.array(city)
    np_casualty[np_casualty == ""] = "0.0"
    np_casualty = np.array(np_casualty, dtype = 'float')
    np_casualty = np.sum(np_casualty, axis = 1)
    index = np.argmax(np_casualty)
    print(int(np_casualty[index]), np_city[index][0], np_city[index][1])
```

```python
## Open and read data file as specified in the question
## Print the required output in given format

import csv
import numpy as np
file = open("terrorismData.csv")
filedata = csv.DictReader(file,skipinitialspace = True)
month = []
year = []
killed = []
wounded =[]
city = []
state = []
terroristgroup =[]
for row in filedata:
    month.append(row["Month"])
    year.append(row["Year"])
    killed.append(row["Killed"])
    wounded.append(row["Wounded"])
    city.append(row["City"])
    terroristgroup.append(row["Group"])
    state.append(row["State"])

killed_arr = np.array(killed)
wounded_arr = np.array(wounded)

killed_arr[killed_arr ==''] = '0.0'
wounded_arr[wounded_arr == ''] = '0.0'

killed_arr = np.array(killed_arr,dtype=float)
wounded_arr = np.array(wounded_arr,dtype=float)

np_month = np.array(month,dtype=int)
bool_arr = (np_month>=5) & (np_month<=7)

np_year = np.array(year,dtype=int)
bool_arr_year = (np_year==1999)

np_state = np.array(state)
bool_state = np_state=="Jammu and Kashmir"

reqd_kills = killed_arr[bool_arr&bool_arr_year]
reqd_wounded = wounded_arr[bool_arr&bool_arr_year]
casualities = killed_arr + wounded_arr
reqd_casualities = casualities[(bool_arr)&(bool_arr_year)&(bool_state)]
index= np.argmax(reqd_casualities)


max_casualities = int(np.max(reqd_casualities))
reqd_city = city[index]
group = terroristgroup[index]

print(max_casualities,reqd_city,group)
```

## Terror City Casualty

```python
## Open and read data file as specified in the question
## Print the required output in given format

import csv
import numpy as np
x={}
with open("terrorismData.csv") as file_:
    file_name=csv.DictReader(file_)
    for i in file_name:
        if i['Country'] !='India'  or i['City']=='Unknown':
            continue
        killed_ = 0
        if i['Killed']!='':
            killed_ = int(float(i['Killed']))
        wounded_ = 0
        if i['Wounded']!='':
            wounded_ = int(float(i['Wounded']))
        causuality_ = killed_ + wounded_
        x[i['City']] = x.get(i['City'],0) + causuality_


y = {k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse=True)}
count = 0
for i in y:
    if count==5:
        break
    print(i, end=' ')
    print(y[i])
    count+=1
```

```python
import csv
import numpy as np

with open('terrorismData.csv', encoding='utf8') as file_obj:
    file_data=csv.DictReader(file_obj, skipinitialspace=True)
    killed=[]
    wounded=[]
    city=[]
    for row in file_data:
        if 'India' in row['Country'] and 'Unknown' not in row['City']:
            city.append(row['City'])
            wounded.append(row['Wounded'])
            killed.append(row['Killed'])
    np_wounded=np.array(wounded)
    np_killed=np.array(killed)
    np_city=np.array(city)
    
    np_killed[np_killed=='']='0.0'
    np_wounded[np_wounded=='']='0.0'
    np_killed=np.array(np_killed, dtype='float')
    np_wounded=np.array(np_wounded, dtype='float')
    np_casuality=np.array(np_wounded+np_killed, dtype='int')
    
    x={}
    for i in np_city:
        if i in x:
            x[np_city]+=np_casuality
        else:
            x[np_city]=np_casuality
y = {k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse=True)}
count = 0
for i in y:
    if count==5:
        break
    print(i, end=' ')
    print(y[i])
    count+=1
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("startup_funding.csv")
df['StartupName'].replace('Flipkart.com','Flipkart',inplace=True)
df['StartupName'].replace('Ola Cabs','Ola',inplace=True)
df['StartupName'].replace('Olacabs','Ola',inplace=True)
df['StartupName'].replace('Oyo Rooms','Oyo',inplace=True)
df['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True)
d = {}
df.dropna(subset=["InvestorsName"],inplace=True)
df.reset_index(drop=True,inplace = True)

for i in df.values:
    startupName = i[2].strip()
    investors = i[6].split(",")
    for inv in investors:
        if inv.strip() in ('','Undisclosed Investors','Undisclosed investors','undisclosed Investors'):
            continue
        v = d.get(inv.strip(),set())
        v.add(startupName)
        d[inv.strip()] = v
invs = []
startups = []
d1 = {}
for k in d.keys():
    d1[k] = len(d[k])
d_sorted = {k: v for k, v in sorted(d1.items(), key=lambda item: item[1],reverse=True)}
count = 0
for k in d_sorted.keys():
    if count==5:
        break
    startups.append(d_sorted[k])
    invs.append(k)
    count+=1

# print(startups)
# print(invs)
for i in range(5):
    print(invs[i],startups[i])
plt.bar(invs,startups)
plt.xticks(rotation=70)
plt.show()
# for i in range(len(invs)):
#     print(startups[i])
```

## 4. top 5 investors who have invested in a different number of startups and their investment type is Crowdfunding or Seed Funding

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("startup_funding.csv")
df['StartupName'].replace('Flipkart.com','Flipkart',inplace=True)
df['StartupName'].replace('Ola Cabs','Ola',inplace=True)
df['StartupName'].replace('Olacabs','Ola',inplace=True)
df['StartupName'].replace('Oyo Rooms','Oyo',inplace=True)
df['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True)
df['InvestmentType'].replace('SeedFunding','Seed Funding',inplace=True)
df['InvestmentType'].replace('Crowd funding','Crowd Funding',inplace=True)
df['InvestmentType'].replace('PrivateEquity','Private Equity',inplace=True)
d = {}
df.dropna(subset=["InvestorsName"],inplace=True)
df.dropna(subset=["InvestmentType"],inplace=True)

df.reset_index(drop=True,inplace = True)

for i in df[(df['InvestmentType']=='Crowd Funding') | (df['InvestmentType']=='Seed Funding')].values:
    startupName = i[2].strip()
    investors = i[6].split(",")
    for inv in investors:
        if inv.strip() in ('','Undisclosed Investors','Undisclosed investors','undisclosed Investors'):
            continue
        v = d.get(inv.strip(),set())
        v.add(startupName)
        d[inv.strip()] = v
invs = []
startups = []
d1 = {}
for k in d.keys():
    d1[k] = len(d[k])
d_sorted = {k: v for k, v in sorted(d1.items(), key=lambda item: item[1],reverse=True)}
count = 0
for k in d_sorted.keys():
    if count==5:
        break
    startups.append(d_sorted[k])
    invs.append(k)
    count+=1

print(startups)
print(invs)
    
plt.bar(invs,startups)
plt.xticks(rotation=70)
plt.show()
```
