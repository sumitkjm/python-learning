```python
## Read input as specified in the question.
## Print output as specified in the question.
from collections import defaultdict
 
# Function to return the number
# of unique triangles
def UniqueTriangles(a, b, c, n):
 
    sides = [None for i in range(n)]
 
    # Map to store the frequency of
    # triangles with same sides
    m = defaultdict(lambda:0)
 
    for i in range(0, n):
 
        # Push all the sides of the current triangle
        sides[i] = (a[i], b[i], c[i])
 
        # Sort the three sides
        sides[i] = tuple(sorted(sides[i]))
 
        # Store the frequency of the sides
        # of the triangle
        m[sides[i]] += 1
 
    # To store the count of unique triangles
    count = 0
    for i in m:
 
        # If current triangle has unique sides
        if m[i] == 1:
            count += 1
 
    return count
 
# n = int(input())
# for i in range(n):
# arr=list(int(i) for i in input().strip().split(' '))
arr = [5 7 6 5 5 7 6 8 2 9 2 3 4 2 4 3]
```

## Plot Job_Year

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("amazon_jobs_dataset.csv")
d = {}
for i in df.values:
    year = i[3].split(",")[1].strip()
    d[year] = d.get(year,0) + 1

years = list(d.keys())
years.sort()
jobs = []
for i in years:
    jobs.append(d[i])

for key in years:
    print(key,end= ' ')
    print(d[key])

plt.plot(years,jobs)
plt.show()
```

## Plot Month_Job

```python
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
df = pd.read_csv("amazon_jobs_dataset.csv")
d = {}
date_r1 = datetime.strptime('June 1, 2011','%B %d, %Y')
date_r2 = datetime.strptime('March 31, 2018','%B %d, %Y')
for i in df.values:
    date = datetime.strptime(i[3],'%B %d, %Y')
    if date>=date_r1 and date<=date_r2:
        month = i[3].split(",")[0].split()[0].strip()
        d[month] = d.get(month,0) + 1
months = []
jobs = []

for i in d:
    months.append(i)
    print(i,end=' ')
    jobs.append(d[i])
    print(d[i])
plt.bar(months,jobs)
plt.show()
```

```python
## Open and read data file as specified in the question
## Print the required output in given format
import numpy as np
import csv
import pandas as pd

file_obj = open("terrorismData.csv",encoding='utf-8')
file_data = csv.DictReader(file_obj,skipinitialspace=True)
month = []
year =[]
killed=[]
wounded=[]
city=[]
group=[]
state=[]
for rows in file_data:
    year.append(rows['Year'])
    month.append(rows['Month'])
    killed.append(rows['Killed'])
    wounded.append(rows['Wounded'])
    city.append(rows['City'])
    group.append(rows['Group'])
    state.append(rows['State'])
Killed = np.array(killed)
Killed[Killed=='']='0.0'
Killed = np.array(Killed,dtype=float)
Wounded = np.array(wounded)
Wounded[Wounded=='']='0.0'
Wounded = np.array(Wounded,dtype=float)
State = np.array(state)
City = np.array(city)
Group = np.array(group)

Casualty = Killed + Wounded
Casualty = Casualty[State=='Jammu and Kashmir']
City = City[State=='Jammu and Kashmir']
Group = Group[State=='Jammu and Kashmir']
Year = np.array(year,dtype=int)
Year = Year[State=='Jammu and Kashmir']
Month = np.array(month,dtype=int)

final_Casualty = Casualty[Year==1999]
final_city = City[Year==1999]
final_group = Group[Year==1999]
Month = Month[State=='Jammu and Kashmir']
Month = Month[Year==1999]

final_Casualty = final_Casualty[(Month>4) & (Month<8)]
final_city = final_city[(Month>4) & (Month<8)]
final_group = final_group[(Month>4) & (Month<8)]
output_city = final_city[final_Casualty==max(final_Casualty)]
output_group = final_group[final_Casualty==max(final_Casualty)]
print(int(max(final_Casualty)),output_city[0],output_group[0])
```
