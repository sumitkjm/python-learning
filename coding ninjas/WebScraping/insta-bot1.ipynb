```python
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(executable_path='C:\\Users\\Sumit\\.wdm\\drivers\\chromedriver\\win32\\95.0.4638.69\\chromedriver')
driver
```

```python
# open instagram page
driver.get("https://www.instagram.com/")
```

```python
# 1. fill username password and login
username = driver.find_element(By.NAME,"username")
password = driver.find_element(By.NAME,"password")
submit = driver.find_element(By.XPATH,'//button[@type="submit"]')
username.send_keys("****")
password.send_keys("****")
submit.click()
```

```python
# click Not now to not save the password in popup window
driver.implicitly_wait(10)
not_now_btn = driver.find_elements(By.XPATH,'//button')
for i in not_now_btn:
    if i.text == 'Not Now':
        i.click()
        break
```

```python
# 2. print search results of "food"
search_text = driver.find_element(By.XPATH,'//input[@aria-label="Search Input"]')
search_text.send_keys("food")
search_result = driver.find_elements(By.XPATH,"//div[@role='none']/a")
for i in search_result:
    href_attribute = i.get_attribute("href")
    if "tags/" in href_attribute:
        continue
    print(i.get_attribute("href").split("/")[3])
```

```python
# 3. print search results of "So Delhi"
search_text = driver.find_element(By.XPATH,'//input[@aria-label="Search Input"]')
search_text.clear()
search_text.send_keys("So Delhi")
search_result1 = driver.find_element(By.XPATH,"//div[@role='none']/a")
search_result1
href_attribute = search_result1.get_attribute("href")
driver.get(href_attribute)
```

```python
# 4. Follow/Unfollow given handle
follow_btn = driver.find_elements(By.XPATH,"//header//button")
following_btn = driver.find_elements(By.XPATH,"//span[@aria-label='Following']")
unfollow_btn = driver.find_elements(By.XPATH,"//span[@aria-label='Following']/../..")
for i in follow_btn:
    if i.text == 'Follow':
        i.click()
        print("Followed")
        break
if len(following_btn) > 0:
    print("Following")
    unfollow_btn[0].click()
    unfollow_conf_btn = driver.find_elements(By.XPATH,"//button")
    for i in unfollow_conf_btn:
        if i.text == 'Unfollow':
            i.click()
            print("Unfollowed")
            break
```

```python
# 5. Like/Unlike posts - Like top 30 posts
like_page_link = driver.find_elements(By.XPATH,"//img[@class='FFVAD']/../../..")
for i in like_page_link[:1]:
    like_link = i.get_attribute("href")
    print(like_link)
    driver.get(like_link)
    driver.implicitly_wait(10)
    like_btn = driver.find_element(By.XPATH,"//svg[@aria-label='Like']/../..")
    print(like_btn)
    driver.back()
```
