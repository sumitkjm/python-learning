```python
import matplotlib.pyplot as plt
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
```

```python
# plt.scatter(year,salary,s=100) # s is size of bubble
plt.scatter(year,salary,s=population,label="bubble") # s is size of bubble
plt.title("Bubble chart")
plt.legend()
```

```python
plt.scatter(year,salary,s=population,label="bubble",c='r',alpha=0.1) # s is size of bubble
plt.title("Bubble chart")
plt.legend()
```

```python
import numpy as np
c = np.arange(len(year))
plt.scatter(year,salary,s=population,label="bubble",c=c,alpha=0.6) # s is size of bubble
plt.title("Bubble chart")
plt.legend()
```

```python
plt.scatter(year,salary,s=population,label="bubble",c=c,alpha=0.6,edgecolor='black') # s is size of bubble
plt.title("Bubble chart")
plt.legend()
```

```python
plt.scatter(year,salary,s=population,label="bubble",c=c,alpha=0.6,edgecolor='black',marker="^") # s is size of bubble
plt.title("Bubble chart")
plt.text(year[0]+0.1,salary[0]+0.1,population[0])
plt.legend()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("amazon_jobs_dataset.csv")
```

```python
d = {}
for i in df.values:
    year = i[3].split(",")[1]
    d[year] = d.get(year,0) + 1
years = list(d.keys())
jobs = list(d.values())
years.sort()
jobs.sort()
print(years)
print(jobs)
plt.plot(years,jobs)

for key in sorted(d):
    print(key,end= ' ')
    print(d[key])
```
