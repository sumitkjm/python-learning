```python
import matplotlib.pyplot as plt
```

```python
sizes = [3,4,6,2]
labels=["A","B","C","D"]
explodes = [0.1,0,0,0]
colors = ["blue","black","pink","orange"]
plt.pie(sizes,explode=explodes, labels=labels,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```

```python
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]

plt.pie(sold,labels=company,autopct="%.1f%%")
plt.show()
```
