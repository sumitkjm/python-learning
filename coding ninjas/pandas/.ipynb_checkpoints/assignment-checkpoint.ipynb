```python
import pandas as pd
df = pd.read_csv("terrorismData.csv")
d = {}
for i in df[(df.State=='Jammu and Kashmir') & (df.Group!='Unknown')].values:
    key = i[6] +":"+i[14]
    d[key] = d.get(key,0) + 1

maxCity = max(d, key= lambda x: d[x])
print(maxCity.split(':')[0],end=' ')
print(d[maxCity],end=' ')
print(maxCity.split(':')[1])
```

```python
import pandas as pd
from datetime import datetime
df = pd.read_csv("terrorismData.csv")
d = {}
compareDate = datetime.strptime('2014-05-26','%Y-%m-%d')
for i in df.values:
    if i[3] == 'India' and i[14] !='Unknown' :
        day = str(i[2])
        if i[2] == 0:
            day = "1"
        date = datetime.strptime(str(i[0]) + "-" + str(i[1]) + "-" + day, '%Y-%m-%d')
        if date>=compareDate:
            d[i[14]] = d.get(i[14],0) + 1
maxT = 0
g = ""
for i in d:
    if d[i] > maxT:
        maxT = d[i]
        g = i
print(maxT)
print(g)
maxGroup = max(d, key= lambda x: d[x])
print(d[maxGroup],end=' ')
print(maxGroup)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("terrorismData.csv")
corridor_states = ["Jharkhand", "Odisha", "Andhra Pradesh", "Chhattisgarh"]
d = {}
d2 = {}
for i in df.values:
    killed = i[10]
    wounded = i[11]
    casuality = killed + wounded
    if i[4] in corridor_states:
        d[i[0]] = d.get(i[0],0) + casuality
    elif i[4] == 'Jammu and Kashmir':
        d2[i[0]] = d2.get(i[0],0) + casuality
    else:
        continue
print(sum(d.values())/len(d.keys()),end=' ')
print(sum(d2.values())/len(d2.keys()))
```

```python
import pandas as pd 
import numpy as np 
df_terrorism=pd.read_csv('terrorismData.csv')
year=len(set(df_terrorism['Year']))
df_terrorism=df_terrorism[df_terrorism['Country']=='India'] 
df_terrorism['Casualty']=df_terrorism['Killed']+df_terrorism['Wounded']
Jammu_state=df_terrorism[df_terrorism['State']=='Jammu and Kashmir'] 
red_state=df_terrorism[(df_terrorism['State']=='Jharkhand')|(df_terrorism['State']=='Odisha') |(df_terrorism['State']=='Andhra Pradesh')|(df_terrorism['State']=='Chhattisgarh')] 
red_casualty=int(np.sum(red_state['Casualty'])) 
Jammu_casualty=int(np.sum(Jammu_state['Casualty'])) 
print(red_casualty//year,Jammu_casualty//year)
```
