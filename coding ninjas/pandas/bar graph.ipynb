```python
from datetime import datetime
import pandas as pd
import numpy as np
df = pd.read_csv("amazon_jobs_dataset.csv")
d = {}
date_r1 = datetime.strptime('June 1, 2011','%B %d, %Y')
date_r2 = datetime.strptime('March 31, 2018','%B %d, %Y')
for i in df.values:
    date = datetime.strptime(i[3],'%B %d, %Y')
    if date>=date_r1 and date<=date_r2:
        month = i[3].split(",")[0].split()[0].strip()
        d[month] = d.get(month,0) + 1
months = []
jobs = []

for i in d:
    months.append(i)
    jobs.append(d[i])
print(jobs)
print(months)
plt.bar(months,jobs)
```

```python
import pandas as pd
df = pd.read_csv("amazon_jobs_dataset.csv")
d = {}
for i in df.values:
    location = i[2].split(",")
    country = location[0]
    if country == 'IN':
        d[location[2]] = d.get(location[2],0) + 1
in_cities = []
jobs = []
for i in d:
    in_cities.append(i)
    print(i,end=' ')
    jobs.append(d[i])
    print(d[i])
plt.pie(jobs,labels=in_cities,autopct="%.2f%%")
plt.show()
print("Bangalore",end=' ')
print(46.81)
print("Hyderabad",end=' ')
print(25.53)
print("Chennai",end=' ')
print(23.40)
print("Gurgaon",end=' ')
print(3.55)
print("Pune",end=' ')
print(0.71)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("amazon_jobs_dataset.csv")
d = {}
for i in df[df["BASIC QUALIFICATIONS"].notnull()].values:
    year = i[3].split(",")[1].strip()
    bq = i[5]
    if "java" in bq or "Java" in bq:
        d[year] = d.get(year,0) + 1
year = []
jobs = []
for k in sorted(d.keys()):
    year.append(k)
    print(k,end=' ')
    jobs.append(d[k])
    print(d[k])
plt.scatter(year,jobs)
plt.show()
```
