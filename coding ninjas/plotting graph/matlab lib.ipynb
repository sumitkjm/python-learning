```python
# plot x^3
import numpy as np
# x = np.array([1,2,3,4])
x = np.arange(0,10,0.1)
y = x**3
y2 = x**2
y3 = x**4
# plt.plot(x,y,color="orange",marker="p")
# plt.ylabel("x^3")
# plt.xlabel("x")
plt.title("Matplotlib demo")
plt.plot(x,y,color="orange",linewidth=2,label="x^3")
plt.plot(x,y2,color="blue",linewidth=2,label="x^2")
plt.plot(x,y3,color="blue",linewidth=2,label="x^4")
plt.axis([0,10,0,1000])
plt.grid()
plt.legend()
plt.text(2,400,"Text",fontsize=12)
```
